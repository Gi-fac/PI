## Exercício 1

A função ``decora`` recebe um _string_ ``texto``, um _string_ ``d`` e um número inteiro positivo ``n``. Em seguida, a função devolve um novo _string_ formado por n repetições de ``d``, seguido por ``texto`` seguido por mais n repetições de ``d``. Por exemplo, ``decora("UFABC", "*", 3)`` produz o _string_ ``***UFABC***``. Nota: em Python, a operação ``um_string*um_natural`` produz o _string_ ``um_string`` repetido ``um_natural`` vezes. Por exemplo, ``"ab"*3`` produz ``"ababab"``.

```
    decoracao = d*n

```
