## Exercício 1

Faça um programa que leia uma sequência de **números inteiros**, **positivos** ou **negativos**, que é encerrada com o número **zero**. O programa deve somar os valores **negativos** e imprimir ao final o valor dessa somatória. O zero final não é considerado parte da sequência. Por exemplo, na sequência: (9, -5, 2, -9, 12, 10, 7,  3, -2, 0)  a resposta procurada é -16. Note que não é necessário armazenar todos os números.

**Entrada**: sequência de números inteiros de tamanho desconhecido, **terminada por zero**.

**Saída**: somatória dos números **negativos** digitados pelo usuário.

```python
soma = 0  # acumulador para somar os negativos
num = int(input("Digite um número inteiro (0 para encerrar): "))  # lê o primeiro número
while num != 0:  # repete enquanto não digitar 0
    if num < 0:  # só soma se for negativo
        soma = soma + num
    num = int(input("Digite um número inteiro (0 para encerrar): "))  # lê o próximo
print("Soma dos negativos:", soma)
```

## Exercício 2

Em um país *X*, a moeda nacional chama-se *conto*. O país aplica uma taxa de alfândega progressiva para bens declarados ao se entrar no país. Bens de valor até 500 (quinhentos) contos são taxados em 5%. Acima de 500 contos, o valor excedente (a quantidade superior a 500) é taxado em 8%. Por exemplo:

- um bem de 100 contos recebe uma taxa de 5 contos.
- um bem de 500 contos recebe uma taxa de 25 contos.
- um bem de 600 contos recebe uma taxa de 33 contos (25 + 8).

Faça um programa que recebe um número positivo do tipo _float_, representando o valor de um bem e em seguida imprime o **valor da taxa aplicada**. Por exemplo, para a entrada 600,00 seu programa deve imprimir 33,00 (não se preocupe com a quantidade de casas decimais ao imprimir).

**Entrada**: um número positivo do tipo float, representando um valor em contos.

**Saída**: o valor da taxa aplicada ao valor de entrada.

```python
valor = float(input("Digite o valor do bem: "))  # lê o valor do bem
if valor <= 500:  # se está dentro da faixa de 5%
    taxa = valor * 0.05  # aplica 5% sobre tudo
else:  # se passa de 500
    taxa = 500 * 0.05 + (valor - 500) * 0.08  # 5% nos primeiros 500 + 8% no excedente
print("Taxa aplicada:", taxa)
```

## Exercício 3

Faça um programa que leia uma sequência de números **inteiros positivos**, que é encerrada com o número **zero**. O programa deve contar e imprimir ao final **quantas vezes**, dentro da sequência, **um número é maior que seu antecessor direto**. O zero final não é considerado parte da sequência. Por exemplo, na sequência: (3, 5, 2, 9, 7 , 6, 4,  4, 8, 0)  a resposta procurada é 3. Note que não é necessário armazenar todos os números, a cada iteração somente é preciso comparar um número com seu antecessor.

**Entrada**: sequência de **números inteiros positivos** de tamanho desconhecido, **terminada por zero**.

**Saída**: **quantidade de vezes** em que **um número é maior que seu antecessor imediato**.


```python
contador = 0  # conta quantas vezes num > anterior
anterior = int(input("Digite um número (0 para encerrar): "))  # lê o primeiro
if anterior != 0:  # só continua se não for 0 de cara
    num = int(input("Digite um número (0 para encerrar): "))  # lê o segundo
    while num != 0:  # repete até digitar 0
        if num > anterior:  # compara atual com anterior
            contador = contador + 1
        anterior = num  # atualiza o anterior
        num = int(input("Digite um número (0 para encerrar): "))  # lê o próximo
print("Quantidade de vezes que um número é maior que seu antecessor:", contador)
```

## Exercício 4

Faça um programa que leia **um número inteiro positivo**, guardando o valor em uma variável chamada ``limiar``. Em seguida, o programa deve ler uma sequência de **números inteiros positivos**, que é **encerrada com o número zero**. A partir do segundo número da sequência, seu programa deve calcular a **diferença entre o número digitado e seu antecessor**:

	diferença = (número atual) - (número anterior)

Seu programa deve imprimir uma mensagem sempre que essa diferença calculada seja **estritamente maior que o valor de limiar**. O zero final não é considerado parte da sequência. Por exemplo, supondo que limiar tenha valor 5 e a sequência fornecida seja (3, 6, 1, 9, 2, 7, 15, 8, 0); seu programa deve imprimir uma mensagem para os números em negrito *9* e *15* da sequência. A mensagem deve dizer: "Limiar ultrapassado para (n, m)", em que *n* deve ser substituído pelo número atual e *m* deve ser substituído pelo número anteriormente digitado. No exemplo anterior, seu programa deve imprimir:

	Limiar ultrapassado para (9, 1)
	Limiar ultrapassado para (15, 7)

Note que **não** é necessário armazenar todos os números, a cada iteração somente é preciso comparar um número com seu antecessor.

```python
limiar = int(input("Digite o valor do limiar: "))  # lê o limiar
anterior = int(input("Digite um número (0 para encerrar): "))  # lê o primeiro número
if anterior != 0:
    num = int(input("Digite um número (0 para encerrar): "))  # lê o segundo
    while num != 0:  # repete até 0
        diferenca = num - anterior  # calcula diferença entre atual e anterior
        if diferenca > limiar:  # se ultrapassou o limiar, imprime
            print(f"Limiar ultrapassado para ({num}, {anterior})")
        anterior = num  # atualiza anterior
        num = int(input("Digite um número (0 para encerrar): "))  # lê próximo
```

## Exercício 5

Em um experimento de laboratório, uma população de bactérias tem inicialmente $N_0$ indivíduos. A cada dia, o número de indivíduos é dado pela equação:

$N_i = t * N_{(i-1)}$

Em que $N_i$ é o número de indivíduos no dia atual, $N_{(i-1)}$ é o número de indivíduos no dia anterior, e $t$ é o fator de crescimento da população. Faça um programa que recebe via teclado dois números reais $N_0$ e $t$ e em seguida calcula o número de indivíduos ao final de cada dia. Pare sua simulação quando o número de indivíduos for maior que $10\cdot N_0$ ou menor que $0.1 \cdot N_0$. Imprima a quantidade de dias que se passaram em sua simulação.

```python
N0 = float(input("Digite o número inicial de indivíduos (N0): "))  # população inicial
t = float(input("Digite o fator de crescimento (t): "))  # taxa de crescimento
N = N0  # população atual começa igual à inicial
dias = 0  # contador de dias
while N <= 10 * N0 and N >= 0.1 * N0:  # para quando ultrapassa os limites
    N = t * N  # aplica o crescimento
    dias = dias + 1  # conta mais um dia
print("Quantidade de dias:", dias)
```

## Exercício 6

Uma sequência de números $a_1, a_2, a_3, \ldots, a_n$ contém uma inversão quando um número não-negativo é sucedido por um número negativo. Por exemplo, na sequência a seguir, as inversões foram assinaladas em negrito: $1, 3, 0, \mathbf{-1}, 6, \mathbf{-3}, -1, 2, 8$.
Crie um programa python que:
1. Leia uma sequência de 8 números digitados pelo usuário.
2. O seu programa deve imprimir se a sequência contém alguma inversão ou não.

Para que a resposta seja considerada, seu programa deve ler os números e iterar sobre eles usando um laço de repetição.


```python
inversao = False  # flag que indica se houve inversão
anterior = int(input("Digite o 1° número: "))  # lê o primeiro
for i in range(7):  # lê os 7 restantes (total = 8)
    num = int(input(f"Digite o {i+2}° número: "))
    if anterior >= 0 and num < 0:  # anterior não-negativo seguido de negativo = inversão
        inversao = True
    anterior = num  # atualiza anterior
if inversao:
    print("A sequência contém inversão.")
else:
    print("A sequência não contém inversão.")
```

## Exercício 7


Crie um programa python que:
1. Leia um número inteiro $n$ digitado pelo usuário.
2. Em seguida, leia uma sequência de números inteiros digitados pelo usuário, terminada por 0 (zero).
3. Seu programa deve imprimir a somatória dos números da sequência **que sejam estritamente maiores que** $n$.

Por exemplo, para $n= 4$ e a sequência $2, 7, 3, 6, 9, 1, 4, 3, 0$, seu programa deve imprimir $22$ (isto é, $7 + 6 + 9$)

```python
n = int(input("Digite o valor de n: "))  # lê o limite n
soma = 0  # acumulador
num = int(input("Digite um número (0 para encerrar): "))  # lê primeiro da sequência
while num != 0:  # repete até 0
    if num > n:  # só soma se for estritamente maior que n
        soma = soma + num
    num = int(input("Digite um número (0 para encerrar): "))  # lê próximo
print("Soma dos números maiores que n:", soma)
```

## Exercício 8

Em uma certa disciplina, um aluno é considerado **aprovado** se obtiver uma **nota final igual ou maior que 5 (cinco)**. Todo aluno que não for aprovado é considerado reprovado. Crie um programa Python que leia uma sequência de números finalizada por $-1$ (esse último valor não é considerado
parte da sequência). Cada número representa a nota final de um aluno. Seu programa deve contar o número de alunos aprovados e reprovados e imprimir as duas contagens ao final.

```python
aprovados = 0  # contador de aprovados
reprovados = 0  # contador de reprovados
nota = float(input("Digite a nota do aluno (-1 para encerrar): "))  # lê primeira nota
while nota != -1:  # repete até digitar -1
    if nota >= 5:  # nota >= 5 é aprovado
        aprovados = aprovados + 1
    else:  # senão é reprovado
        reprovados = reprovados + 1
    nota = float(input("Digite a nota do aluno (-1 para encerrar): "))  # lê próxima
print("Aprovados:", aprovados)
print("Reprovados:", reprovados)
```

## Exercício 9

Crie um programa Python que leia 10 números digitados pelo usuário a partir do teclado. O seu programa deve imprimir:
1. A média $m$ dos números digitados pelo usuário.
2. Todos os valores digitados que são menores que $0.5m$.
3. Todos os valores digitados que são maiores que $0.75m$.


```python
numeros = []  # lista para guardar os 10 números (precisa comparar depois)
for i in range(10):  # lê 10 números
    num = float(input(f"Digite o {i+1}° número: "))
    numeros.append(num)  # adiciona na lista
m = sum(numeros) / 10  # calcula a média
print("Média:", m)
print("Valores menores que 0.5m:")
for num in numeros:  # percorre a lista
    if num < 0.5 * m:  # menor que metade da média
        print(num)
print("Valores maiores que 0.75m:")
for num in numeros:  # percorre a lista de novo
    if num > 0.75 * m:  # maior que 75% da média
        print(num)
```

## Exercício 10

Uma empresa produz peças com tamanho médio de  100mm, com desvio máximo aceitável de 5mm para mais ou para menos. Peças fora desse padrão são descartadas. Crie um programa Python que leia uma sequência de números finalizada por $-1$ (esse último valor não é considerado parte da sequência). Cada número representa o tamanho de uma peça dentro de um lote. Seu programa deve contar o número de peças total e o número de peças fora do padrão e imprimir a porcentagem de peças reprovadas.

```python
total = 0  # conta todas as peças
fora_padrao = 0  # conta as fora do padrão
tamanho = float(input("Digite o tamanho da peça (-1 para encerrar): "))  # lê primeira
while tamanho != -1:  # repete até -1
    total = total + 1  # conta a peça
    if tamanho < 95 or tamanho > 105:  # fora do intervalo [95, 105]
        fora_padrao = fora_padrao + 1
    tamanho = float(input("Digite o tamanho da peça (-1 para encerrar): "))  # lê próxima
if total > 0:  # evita divisão por zero
    porcentagem = (fora_padrao / total) * 100
    print("Porcentagem de peças reprovadas:", porcentagem, "%")
else:
    print("Nenhuma peça foi inserida.")
```

## Exercício 11

Crie um programa Python que leia 10 números digitados pelo usuário a partir do teclado. O seu programa deve imprimir:
1. O valor mínimo $m$ dos números digitados pelo usuário.
2. A quantidade de números digitados que estão contidos no intervalo $[m, 2m]$.


```python
numeros = []  # lista para guardar os 10 números
for i in range(10):  # lê 10 números
    num = float(input(f"Digite o {i+1}° número: "))
    numeros.append(num)
m = numeros[0]  # assume o primeiro como mínimo
for num in numeros:  # percorre para achar o menor
    if num < m:
        m = num
print("Valor mínimo:", m)
contador = 0  # conta quantos estão no intervalo [m, 2m]
for num in numeros:
    if num >= m and num <= 2 * m:  # verifica se está no intervalo
        contador = contador + 1
print("Quantidade no intervalo [m, 2m]:", contador)
```

## Exercício 12

Faça um programa que leia uma sequência de números inteiros, que é encerrada com o número zero. O programa deve calcular e apresentar a soma de todos os valores informados que forem **menores que zero**, ou então imprimir uma mensagem quando não ocorreram valores menores que zero.

```python
soma = 0  # acumulador dos negativos
tem_negativo = False  # flag para saber se apareceu algum negativo
num = int(input("Digite um número inteiro (0 para encerrar): "))  # lê primeiro
while num != 0:  # repete até 0
    if num < 0:  # se for negativo, soma e marca a flag
        soma = soma + num
        tem_negativo = True
    num = int(input("Digite um número inteiro (0 para encerrar): "))  # lê próximo
if tem_negativo:  # se teve algum negativo, mostra a soma
    print("Soma dos valores negativos:", soma)
else:  # se não teve, mostra mensagem
    print("Não ocorreram valores menores que zero.")
```
