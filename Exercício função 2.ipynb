## Exercício 1

Crie um programa python que:
1. Leia três números inteiros digitados pelo usuário.
2. Imprima o valor da **soma** *dos dois maiores números* **dividida** *pelo menor número*.

Use a célula abaixo para inserir sua resposta.

A leitura de valores pode ser feita com a função input, como no exemplo abaixo:

```python

```

````python
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))
c = int(input("Digite o terceiro número: "))

if a <= b and a <= c:
    menor = a
    soma_maiores = b + c
elif b <= a and b <= c:
    menor = b
    soma_maiores = a + c
else:
    menor = c
    soma_maiores = a + b

resultado = soma_maiores / menor
print("Resultado:", resultado)
````


````python
def dia_semana(dia, mes, ano):
    a0 = ano - (14 - mes) // 12
    x = a0 + a0 // 4 - a0 // 100 + a0 // 400
    m0 = mes + 12 * ((14 - mes) // 12) - 2
    d0 = (dia + x + (31 * m0) // 12) % 7
    return d0
````

````python
# o código abaixo testa a função dia_semana;
# se sua função estiver correta, o código
# abaixo deve executar sem erro
if dia_semana(13, 6, 2021) != 0:
    print("Erro, deveria ser domingo.")
if dia_semana(1, 5, 2021) != 6:
    print("Erro, deveria ser sábado.")
if dia_semana(1, 12, 2020) != 2:
    print("Erro, deveria ser terça-feira.")
````

## Exercício 3

Em um exercício anterior, você criou código para verificar se um ano é bissexto ou não. Transforme esse código em uma função, que recebe um ano como um valor numérico e devolve o valor Booleano ``True``caso o ano seja bissexto e devolve ``False`` caso contrário.
Ao criar o código, reflita sobre os seguintes pontos:
- Qual seria um nome adequado para essa função, que expresse claramente seu propósito?
- Quais são os valores que a função deve aceitar? O que deveria acontecer caso o valor passado como argumento seja negativo? Ou fracionário?

````python
def eh_bissexto(ano):
    if ano % 400 == 0:
        return True
    elif ano % 100 == 0:
        return False
    elif ano % 4 == 0:
        return True
    else:
        return False
````

Agora escreva um trecho de código para testar sua função. Você deve testar com valores que sejam anos bissextos e também que não sejam. Reflita sobre o algoritmo de verificação de ano bissexto que você implementou. Quais valores você deveria usar para testar sua função, de maneira que todos os cenários  do algoritmo sejam testados?

````python
# Testes da função eh_bissexto
print(eh_bissexto(2000))  # True (divisível por 400)
print(eh_bissexto(1900))  # False (divisível por 100, mas não por 400)
print(eh_bissexto(2024))  # True (divisível por 4, mas não por 100)
print(eh_bissexto(2023))  # False (não divisível por 4)
````

## Exercício 5

O cálculo de *juros compostos* envolve um valor monetário inicial, denominado *capital*, uma taxa de juros *j* por período e o tempo total *t* transcorrido. Ao final desse tempo, o valor corrigido pelos juros, denominado *montante* é determinado pela fórmula:

$$\text{montante} =  \text{capital}(1 + j)^{t}$$

Já no caso de *juros simples*, o valor da taxa de juros é aplicada sempre ao valor do capital inicial, resultando na fórmula:

$$\text{montante} =  \text{capital}(1 + jt)$$

1. Crie uma funçao ``juro_composto`` que recebe um valor inicial monetário, um valor de taxa (mensal) de juros e uma quantidade de meses e devolve o valor do montante final. Use a fórmula de juros compostos.
2. Use a função acima para calcular o montante final para um empréstimo de R\$ 1000,00, com taxa de 0.5\% ao mês e tempo total de 12 meses.
3. Calcule a taxa de juros simples que resultaria no mesmo montante calculado no item anterior, nas mesmas condições (valor  inicial de mil reais,  tempo total de 12 meses). Você deve resolver esta questão usando código Python.

Ao resolver a questão, reflita sobre o seguinte ponto: valores monetários não se comportam exatamente como números reais. Por exemplo, não é possível ir ao banco e sacar R\$ 1,3958672; os valores monetários usuais são limitados a duas casas decimais. Como você acha que o seu programa deveria lidar com essa questão? Se for necessário "arredondar" valores dentro do seu programa, quando isso deveria ser feito?

````python
# 1. Função juro_composto
def juro_composto(capital, j, t):
    montante = capital * (1 + j) ** t
    return montante

# 2. Calcular montante para R$1000, 0.5% ao mês, 12 meses
montante = juro_composto(1000, 0.005, 12)
print("Montante com juros compostos:", round(montante, 2))

# 3. Calcular taxa de juros simples equivalente
# montante = capital * (1 + j_simples * t)
# j_simples = (montante / capital - 1) / t
j_simples = (montante / 1000 - 1) / 12
print("Taxa de juros simples equivalente:", round(j_simples, 6))
````
