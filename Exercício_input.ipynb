A função ``input``
=============================

Em ``Python``, a função ``input`` permite capturar dados digitados pelo usuário. A função tem um argumento, chamado ``prompt``, que é mostrado para o usuário do programa como uma mensagem. Em geral, usamos o prompt para instruir o usuário, como no exemplo abaixo:

```
# o argumento prompt é uma mensagem textual mostrada para o usuário
nome = input("Digite o seu nome: ")
print("Bom dia, ", nome)
```

Quando a função ``input`` é chamada, a execução do programa será interrompida até que o usuário digite alguma sequência de teclas, terminando com a tecla ``ENTER``.

A função ``input`` devolve a informação digitada pelo usuário em formato de texto (usando o tipo ``str``). Se quisermos usar a informação como outro tipo de dado (por exemplo, numérico), é necessário fazer uma _conversão de tipo de dado_.

Execute o código abaixo e explique por que a execução não é finalizada, isto é, o programa termina com um erro:


````python
idade = input("Qual é a sua idade (em anos)?")
ano_nascimento = 2025 - idade
print("Você nasceu em ", ano_nascimento)
````

As funções ``int`` e ``float`` podem ser usadas para converter texto (``str``) em tipos núméricos:

````python
numero_jogos = int("4")
numero_pontos = float("3.5")
print("Jogos = ", numero_jogos, ", tipo = ", type(numero_jogos))
print("Pontos = ", numero_pontos, ", tipo = ", type(numero_pontos))
````

Note que a informação textual precisa ser compatível com o tipo de dado de destino. Por exemplo, não podemos converter para inteiro um _string_ que contém um número com parte decimal:

````python
int("7.2")
````

O comportamento é diferente quando ``int`` é usada para converter um valor numérico para inteiro. Veja o exemplo abaixo:

````python
x = int(3.2)
print(x)
print(type(x))
````
