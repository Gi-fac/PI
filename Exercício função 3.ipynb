## ExercĂ­cio 1

Crie uma funĂ§ĂŁo ``centraliza(texto, num_colunas)`` que recebe um _string_ texto e um nĂșmero inteiro positivo ``num_colunas`` e produz uma versĂŁo centralizada do texto, usando espaĂ§os, considerando uma largura total de ``num_colunas``. Por exemplo, ``centraliza("UFABC", 11)`` deve produzir o _string_ ``"   UFABC   "`` (isto Ă©, 3 espaĂ§os, seguidos por ``"UFABC"``, seguidos por mais 3 espaĂ§os, totalizando 11 caracteres).
Lembre-se que:
- a funĂ§ĂŁo ``len(meu_string)`` devolve o nĂșmero de caracteres do argumento ``meu_string``.
- a operaĂ§ĂŁo ``meu_string*n`` produz um novo _string_ composto por n repetiĂ§Ă”es de ``meu_string``.

```python
def centraliza(texto, num_colunas):
    espacos_total = num_colunas - len(texto)
    espacos_esq = espacos_total // 2
    espacos_dir = espacos_total - espacos_esq
    return " " * espacos_esq + texto + " " * espacos_dir

# teste
print(repr(centraliza("UFABC", 11)))  # '   UFABC   '
print(repr(centraliza("OI", 6)))  # '  OI  '
```

## ExercĂ­cio 2

Crie uma funĂ§ĂŁo ``combina_string(a, b)`` que recebe dois _strings_ ``a`` e ``b`` e devolve a concatenaĂ§ĂŁo _menor+maior+menor_ em que _menor_ Ă© o _string_ mais curto entre ``a`` e ``b`` e _maior_ Ă© o _string_ mais longo. Por exemplo:
- ``combina_string("UFABC", "~")`` devolve ``~UFABC~``.
- ``combina_string("+++", "HOJE")`` devolve ``+++HOJE+++``.

```python
def combina_string(a, b):
    if len(a) <= len(b):
        menor = a
        maior = b
    else:
        menor = b
        maior = a
    return menor + maior + menor

# teste
print(combina_string("UFABC", "~"))  # ~UFABC~
print(combina_string("+++", "HOJE"))  # +++HOJE+++
```

## ExercĂ­cio 3

Crie uma funĂ§ĂŁo ``preenche_esq(num, largura)`` que recebe dois nĂșmero inteiros positivos ``num`` e ``largura``. A funĂ§ĂŁo deve devolver uma representaĂ§ĂŁo como _string_ de ``num`` preenchida com zeros Ă  esquerda atĂ© atingir a largura determinada. Caso o nĂșmero de dĂ­gitos de ``num`` seja maior que ``largura``, a funĂ§ĂŁo deve devolver apenas ``num`` convertido para _string_, sem preenchimento. Por exemplo:
- ``preenche_esq(10, 4)`` devolve "0010".
- ``preenche_esq(123, 3)`` devolve "123".
- ``preenche_esq(1035, 3)`` devolve "1035".

Pense em primeiro converter ``num`` para _string_. VocĂȘ pode usar a funĂ§ĂŁo ``str`` para isso, isto Ă©, ``str(num)``. Em seguida, calcule o tamanho da representaĂ§ĂŁo de ``num`` como _string_. Com isso, vocĂȘ pode determinar quantos zeros devem ser adicionados.

```python
def preenche_esq(num, largura):
    texto = str(num)
    zeros = largura - len(texto)
    if zeros > 0:
        return "0" * zeros + texto
    else:
        return texto
```

```python
# testes
print(preenche_esq(10, 4))  # 0010
print(preenche_esq(123, 3))  # 123
print(preenche_esq(1035, 3))  # 1035
```
