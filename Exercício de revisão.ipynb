## Exercício 1

Crie uma função ``repete_string`` que recebe como argumentos um _string_ ``texto`` e um número inteiro positivo ``n`` e devolve um _string_ contendo ``texto`` repetido ``n`` vezes. Caso ``n`` seja negativo, sua função deve devolver o valor ``None``.

```python
def repete_string(texto, n):
    if n < 0:
        return None
    resultado = ""
    for i in range(n):
        resultado = resultado + texto
    return resultado
```

Use a célula abaixo para adicionar mais alguns testes para sua função, usando o código fornecido como modelo.

```python
resultado = repete_string("Oi", 1)
if resultado != "Oi":
    print("Erro: esperava Oi, recebeu: ", resultado)

resultado = repete_string("ab", 3)
if resultado != "ababab":
    print("Erro: esperava ababab, recebeu: ", resultado)

resultado = repete_string("x", -1)
if resultado != None:
    print("Erro: esperava None, recebeu: ", resultado)

resultado = repete_string("hi", 0)
if resultado != "":
    print("Erro: esperava string vazio, recebeu: ", resultado)
```

## Exercício 2

Crie uma função chamada ``reverte`` que recebe como argumento um _string_ ``texto`` e devolve o _string_ de entrada revertido, isto é, caso o argumento seja _abcde_ o valor devolvido será _edcba_. Você deve usar um laço de repetição para reverter o _string_, não é permitido usar funções já existentes ou apenas truques de indexação.

```python
def reverte(texto):
    resultado = ""
    for i in range(len(texto) - 1, -1, -1):
        resultado = resultado + texto[i]
    return resultado
```

Use a célula abaixo para testar sua função.

```python
print(reverte("abcde"))  # edcba
print(reverte("Python"))  # nohtyP
print(reverte(""))  # (vazio)
```

## Exercício 3

Crie uma função ``intercala`` que recebe dois _strings_ ``texto`` e ``car`` e devolve um novo _string_ obtido ao intercalar cada caractere de ``texto`` com o conteúdo de ``car``. Por exemplo, ``intercala("abc", "*")`` deve devolver o _string_ ``"a*b*c"``; ``intercala("HOJE", "_+_")`` deve devolver ``"H_+_O_+_J_+_E"``.

```python
def intercala(texto, car):
    resultado = ""
    for i in range(len(texto)):
        resultado = resultado + texto[i]
        if i < len(texto) - 1:
            resultado = resultado + car
    return resultado
```

Use a célula abaixo para testar sua função.

```python
print(intercala("abc", "*"))  # a*b*c
print(intercala("HOJE", "_+_"))  # H_+_O_+_J_+_E
```

## Exercício 4

Em Python, podemos testar se um valor faz parte de uma lista usando o operador ``in``, como nos exemplos abaixo:

```python
print(2 in [1, 2, 3])
print(4 in [1, 2, 3])
print("azul" in ["verde", "vermelho", "azul"])
print("Japão" in ["Argentina", "Brasil", "Paraguai", "Uruguai"])
```

Crie uma função ``contida(elementos, colecao)`` que avalia se a lista ``elementos`` está contida na lista ``colecao``. Por exemplo, ``contida([1, 3], [1, 2, 3] )`` deve devolver ``True``; ``contida([2, 4], [1, 2, 3])`` deve devolver ``False``; ``contida([2, 4], [2, 4, 6])`` deve devolver ``True``, ``contida([2, 4, 6], [2, 4])`` deve devolver ``False``. Caso ``elementos`` seja uma lista vazia (isto é, de comprimento _zero_) sua função deve devolver ``True``.

```python
def contida(elementos, colecao):
    for item in elementos:
        if item not in colecao:
            return False
    return True
```

Execute os testes abaixo para a sua função:

```python
if not contida([1, 3], [1, 2, 3] ):
    print('Erro: [1, 3] está contida em [1, 2, 3]')
if contida([2, 4], [1, 2, 3]):
    print('Erro [2, 4] não está contida em [1, 2, 3]')
if not contida([2, 4], [2, 4, 6]):
    print('Erro [2, 4] está contida em [2, 4, 6]')
if contida([2, 4, 6], [2, 4]):
    print('Erro [2, 4, 6] não está contida em [2, 4]')
```

## Exercício 5

Crie uma função ``conta_pares`` que recebe uma lista de números naturais e devolve a quantidade de números pares na lista.

```python
def conta_pares(lista):
    contador = 0
    for num in lista:
        if num % 2 == 0:
            contador = contador + 1
    return contador
```

Use a célula abaixo para testar sua função:

```python
print(conta_pares([1, 2, 3, 4, 5, 6]))  # 3
print(conta_pares([1, 3, 5]))  # 0
print(conta_pares([2, 4, 6]))  # 3
```

## Exercício 7

Nós definimos a _amplitude_ de um conjunto de números como a diferença entre o maior número e o menor número do conjunto. Escreve uma função ``amplitude`` que recebe uma lista de números reais e devolve a amplitude do conjunto de números da lista. Caso a lista seja vazia (comprimento zero), sua função deve devolver o valor ``None``.

```python
def amplitude(lista):
    if len(lista) == 0:
        return None
    maior = lista[0]
    menor = lista[0]
    for num in lista:
        if num > maior:
            maior = num
        if num < menor:
            menor = num
    return maior - menor
```

Use a célula abaixo para testar sua função.

```python
print(amplitude([1, 5, 3, 9, 2]))  # 8
print(amplitude([4, 4, 4]))  # 0
print(amplitude([]))  # None
```

## Exercício 8

Dada uma lista _A_ contendo _r_ números ``[a1, a2, ..., ar]``, definimos a versão _centralizada_ da lista _A_ como a lista _A'_ obtida ao subtrair a _média_ de _A_ de cada elemento de _A_. Isto é, _A'_ é a lista ``[a1 - m, a2 - m, ..., ar - m]`` em que _m_ é a média aritmética dos números contidos em _A_.

Escreva uma função ``centraliza`` que recebe uma lista de números reais e devolve a versão _centralizada_ da lista de entrada (você deve criar uma nova lista, a função não deve alterar a lista original).


```python
def centraliza(lista):
    m = sum(lista) / len(lista)
    nova_lista = []
    for num in lista:
        nova_lista.append(num - m)
    return nova_lista
```

Use a célula abaixo para testar sua função:

```python
print(centraliza([1, 2, 3, 4, 5]))  # [-2.0, -1.0, 0.0, 1.0, 2.0]
print(centraliza([10, 20, 30]))  # [-10.0, 0.0, 10.0]
```
