Considere o trecho de código abaixo, em que criamos duas variáveis `x` e `y` e atribuímos valores.

```python

```

No entanto, as linhas de código foram embaralhadas (cada linha individual faz parte da solução correta, mas a ordem das linhas está errada). Ordene as linhas abaixo para implementar a inversão entre `x` e `y`.

```python
y = temp
x = y
temp = x
```

Você pode testar sua solução na célula de código abaixo:

````python
x = input("Digite o primeiro valor: ")
y = input("Digite o segundo valor: ")
print("Inicial:")
print("x = ", x, "e y = ", y)
# copie e cole aqui as linhas fornecidas acima
# na ordem correta
temp = x
x=y
y=temp
# fim do trecho para copiar e colar
print("Final:")
print("x = ", x, "e y = ", y)
````

### Exercício 4

Neste exercício você deve calcular e imprimir a quantidade de moedas que seriam necessárias para um determinado valor de troco.

Vamos guardar um valor inteiro entre $1$ e $99$ em uma variável chamada `troco`:
```python
troco = input("Digite um valor inteiro entre 1 e 99: ")
```

Essa variável representa uma quantidade de centavos de real que queremos representar com uma certa quantidade de moedas. Nós temos disponíveis moedas de $1$, $5$, $10$, $25$ e $50$ centavos. Escreva na célula abaixo o código para calcular e imprimir a quantidade de cada moeda usada para representar o troco. Por exemplo, para `troco = 35`, o seu programa deve imprimir algo como:

```
0 moeda(s) de 50 centavos.
1 moeda(s) de 25 centavos.
1 moeda(s) de 10 centavos.
0 moeda(s) de 5 centavos.
0 moeda(s) de 1 centavo.
```

Este problema tem uma solução trivial, que é fornecer o troco somente em moedas de um centavo. Tente produzir uma solução que minimize o número total de moedas, ou pelo menos use uma diversidade de tipos de moeda.

Use a célula abaixo para entrar com seu código:

````python
troco = (int(input("Digite um valor inteiro entre 1 e 99: ")))
print(troco // 50, "moedas de 50")
troco = troco % 50
print(troco // 25, "moedas de 25")
troco = troco % 25
print(troco // 10, "moedas de 10")
troco = troco % 10
print(troco // 5, "moedas de 5")
troco = troco % 5
print(troco // 1, "moedas de 1")
troco = troco % 1
````

### Exercício 5

Queremos escrever um programa que recebe uma data, representada por uma
sequência de três números equivalentes ao dia ($d$), mês ($m$) e ano ($a$).
Em seguida, o programa deve imprimir um número, correspondente ao dia
da semana dessa data (0 para domingo, 1, para segunda-feira, ..., 5 sexta-feira,
6 para sábado).

Vamos usar as seguintes formulas, adequadas para
o calendário gregoriano ($d_0$ é o dia da semana calculado, domingo
é equivalente a zero, segunda-feira é equivalente a um e assim por diante):

\begin{equation}
  a_0 = a - (14 - m) / 12
\end{equation}
\begin{equation}
  x = a_0 + a_0/4 - a_0/100 + a_0/400
\end{equation}
\begin{equation}
m_0 = m + 12 * ((14 - m) / 12) - 2
\end{equation}
\begin{equation}
d_0 = (d + x + (31*m_0)/ 12) \mod 7
\end{equation}

Escreva na célula abaixo o código para calcular o dia da semana, transcrevendo as equações para a linguagem Python. Precisamos lembrar que a operação de divisão inteira é representada por `//` e que a operação *mod* (resto da divisão) é representada em python pelo operador `%`.

````python
# dia
d = 31
# mes
m = 5
# ano
a = 2021


a=a-(14-m)//12
m=m+12*((14-m)//12)-2
d=d+((m+12*((14-m)//12)-2)+(a-(14-m)//12)%7)%7


dia_da_semana = ["domingo", "segunda", "terca", "quarta", "quinta", "sexta", "sabado"]  

if(d%7==0):
    print("O dia da semana é: ", dia_da_semana[0])
elif(d%7==1):
    print("O dia da semana é:", dia_da_semana[1])
elif(d%7==2):
    print("O dia da semana é:", dia_da_semana[2])
````
