## Exercício 1

Faça um programa python que:
1. Leia 5 números **inteiros** digitados pelo usuário.
2. Imprima a somatória dos números _positivos_ (maiores do que zero) digitados pelo usuário.

Você **deve** usar um laço de repetição para a leitura dos números. Use a célula abaixo para inserir sua resposta.


```python
soma_positivos = 0
for i in range(5):
    numero = int(input(f"Digite o {i+1}° número inteiro: "))
    if numero > 0:
        soma_positivos = soma_positivos + numero
print("A soma dos números positivos é:", soma_positivos)
```

## Exercício 2

Dados três números reais *positivos* e *não-nulos* $x$, $y$ e $z$, dizemos que existe uma _discrepância_ se o maior número é maior que 10 vezes a média dos outros dois.

Por exemplo, para $x = 2$, $y = 17$ e $z = 1$, ocorre uma discrepância porque $17 > 10(1 + 2)/2 = 15$. Para $x = 4$, $y = 8$ e $z = 6$, não ocorre discrepância pois $8 < 10(4 + 6)/2 = 50$

Crie um programa que recebe três números *reais e positivos* $x$, $y$ e $z$ digitados pelo usuário e depois imprime uma mensagem dizendo se existe ou não _discrepância_ entre eles.

- Você **não pode** usar as funções do Python ``min``e ``max`` para resolver essa questão, é necessário comparar os valores diretamente.
- O usuário pode digitar os valores em qualquer ordem (por exemplo, você **não pode** supor que o primeiro valor é o menor e o terceiro é o maior).
- Suponha que o usuário vai digitar apenas número reais positivos, você não precisa tratar possíveis erros de entrada de dados.


```python
x = float(input("Digite x: "))
y = float(input("Digite y: "))
z = float(input("Digite z: "))

if x >= y and x >= z:
    maior = x
    soma = y + z
elif y >= x and y >= z:
    maior = y
    soma = x + z
else:
    maior = z
    soma = x + y

limite = 10 * (soma / 2)

if maior > limite:
    print("Existe discrepância.")
else:
    print("Não existe discrepância.")
```

Crie uma função ``charmosos(x, y, z)`` que recebe três números inteiros **maiores que 1** e devole ``True`` caso os números sejam *charmosos* e ``False``caso contrário. Note que os argumentos ``(x, y, z)`` podem ser fornecidos em qualquer ordem. **Você não pode supor que** $x \leq y \leq z$.

A célula abaixo contem dois casos de teste para a função. Escreva pelo menos mais dois casos de teste, usando o mesmo padrão dos casos fornecidos.

```python
def charmosos(x, y, z):
    # encontrar o menor (a), o do meio (b) e o maior (c)
    if x <= y and x <= z:
        a = x
        if y <= z:
            b = y
            c = z
        else:
            b = z
            c = y
    elif y <= x and y <= z:
        a = y
        if x <= z:
            b = x
            c = z
        else:
            b = z
            c = x
    else:
        a = z
        if x <= y:
            b = x
            c = y
        else:
            b = y
            c = x
    return b % a == c % a

# o código abaixo testa a função
teste_falhou = False
# caso de teste 1
a = 2; b = 2; c = 2
if not charmosos(a, b, c):
    print("Falha para:", a, b, c, "; esperava: True")
    teste_falhou = True
# caso de teste 2
a = 3; b = 4; c = 8
if charmosos(a, b, c):
    print("Falha para:", a, b, c, "; esperava: False")
    teste_falhou = True
# caso de teste 3
a = 6; b = 2; c = 4
if not charmosos(a, b, c):
    print("Falha para:", a, b, c, "; esperava: True")
    teste_falhou = True
# caso de teste 4
a = 5; b = 7; c = 3
if charmosos(a, b, c):
    print("Falha para:", a, b, c, "; esperava: False")
    teste_falhou = True

# não altere o código a partir daqui
if not teste_falhou:
    print("Todos os testes passaram.")
```
