 # Exercício 1

Em um jogo, o objetivo é montar uma torre de blocos de madeira de altura $X$ centímetros. Existem dois tipos de blocos: (i) de 5 cm de altura e (ii) de 1 cm de altura. Faça um programa que recebe do usuário: a altura $X$ da torre pretendida; a quantidade disponível de blocos maiores (5 cm) e a quantidade disponível de blocos menores (1 cm). Seu programa deve imprimir uma resposta dizendo se é possível (ou não) construir uma torre da altura desejada somente com os blocos disponíveis. Imagine que todos os blocos têm base quadrada de mesma medida e que não é possível girar os blocos: a dimensão ''altura'' precisa ficar alinhada com o eixo vertical.

Exemplos:
- Altura = 10cm; 1 bloco de 5cm; 4 blocos de 1cm: **não** é possível construir a torre.
- Altura = 12cm; 2 blocos de 5cm; 7 blocos de 1cm: é possível construir a torre.
- Altura = 8cm; 3 blocos de 5cm; 2 blocos de 1cm: **não** é possível construir a torre.


# Exercício 2

Escreva um programa que recebe três números inteiros positivos fornecidos pelo usuário e que devolve a soma dos valores de entrada que **não são** divisíveis por $3$.


Exemplos:
- (1, 7, 2): devolve 10.
- (2, 9, 4): devolve 6.
- (3, 6, 18): devolve 0.

# Exercício 3

Você deseja carregar uma caixa para transporte com $N$ litros de água. Existem disponíveis garrafas de 2 litros, 1 litro e 500ml. Suponha que você sempre vai preferir uma garrafa de maior capacidade antes de carregar uma garrafa de menor capacidade. Por exemplo, para carregar 5 litros, você escolherá 2 garrafas de 2 litros e 1 garrafa de 1 litro, caso possível. Escreva um programa que recebe do usuário um número inteiro $N$ e as quantidades de garrafas de 2 litros, 1 litro e 500ml. Seu programa deve imprimir na saída: (i) se é possível ou não carregar a quantidade desejada de litros com as garrafas disponíveis e (ii) a quantidade de cada tipo de garrafa para carregar na caixa.

# Exercício 4

Dados três números $x$, $y$ e $z$, dizemos que os números são _desbalanceados_ se a diferença $max(x, y, z) - min(x, y, z)$ é pelo menos 10 vezes maior que o menor valor entre $max(x, y, z) - med(x, y, z)$ e $med(x, y, z) - min(x, y, z)$, em que $med$ indica o valor mediano. Faça um programa que recebe três números reais e imprime uma saída dizendo se os valores são desbalanceados ou não. Você não pode usar as funções $max$ e $min$ do Python neste exercício.
